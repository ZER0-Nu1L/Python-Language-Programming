## 静电学

### 静电力

真空电导率：$\varepsilon_0=8.854\times 10^{-12} C^2/(N\cdot m^2)$


#### 带电直导线外一点

$$
E_y=\frac{\lambda}{4\pi \varepsilon_o a}(\cos\theta_1 - \cos \theta_2)
$$

$$
E_x=\frac{\lambda}{4\pi \varepsilon_0 a} (\sin\theta_2-\sin\theta_1)
$$

#### 带电圆环外一点

$$
E=\frac{1}{4\pi\varepsilon_0}\frac{qx}{(R^2+x^2)^\frac{3}{2}}
$$

#### 无限大平面

$$
E=\frac{\sigma}{2\varepsilon_0}
$$

> 这边的分子是**面密度**
> 可以发现无论距离板面的距离有多远，电场强度都是一样大的  
> 推广到两块平行板间的场强，两板间距远小于板面度量  
> 内部场强：  
> $E=\frac{\sigma}{\varepsilon_0}$  
> 外部场强：  
> $E=0$  
> 这里因为两块板的间距很小，所以可以忽略一些小的偏差

#### 带点半球壳

$$
q=\sigma2\pi R\sin \theta d\theta
$$
> 之后连同其他量带入带点圆环的公式求解

&emsp;**关键是从原来的电荷密度中获得单位长度带电量的表达式**  
&emsp;**然后表示出作用力，获得被积函数，之后只需要做积分运算即可**

### 高斯定理

$$
\Phi_e=\frac{\sum{q_内}}{\varepsilon_0}
$$
&emsp;体现了电场是有源的，内部电荷释放出的电场线一定会从面内穿出来，而传入面内的电场线一定会穿出去
> 对于封闭曲面当电场线从面内穿出时$d\Phi_e$为正; 当电场线由面外穿入时$d\Phi_e$为负（**穿出为正，穿入为负**）

#### 高斯定理的微分形式

$$ \nabla \cdot B = \frac{\rho}{\varepsilon_0} $$

> $\rho$表示了电荷的密度分布，可以求解电荷的分布情况

### 环路定理

&emsp;静电力做功的大小只与终末位置有关，与运动路径无关——静电力是保守力，静电场是保守场  
&emsp;静电场中电场强度的环路积分为0  

$$ \oint_L \vec{E} \cdot d\vec{l}=0 $$

&emsp;表示了静电场是**无旋场**，**旋度为0**（在每一个微小的闭合环路上的环路积分都为0）

### 电势

$$ U_a=\int_a^{零势点}\vec{E}\cdot d \vec{l}$$

&emsp;电势是相对的，相对于0势能点的一个量

#### 点电荷电势计算公式

$$ U=\frac{q}{4\pi \varepsilon_0 r} $$

> 在计算电势的时候，就可以利用**点电荷的电势表达式+电势叠加原理**来计算一点的电势
> 对于连续分布的带电体，可将其分割为无数多电荷元$dq$，每个电荷元$dq$当作点电荷，其电势为：
> $$ dU=\frac{dq}{4 \pi \varepsilon_0 r} $$
> 之后对于整个体做积分即可

#### 圆环轴线上点的电势

$$ U=\frac{q}{4 \pi \varepsilon_0 r} $$

> $r$表示轴线上点到圆环上点的最小距离
> $q$表示圆环带的电荷量
> 因为轴线上一点到圆环上各点的**距离都是一样的**，所以可以合并为**同一点**，看作到一**点电荷**的电势

#### 圆盘轴线上点的电势

$$ U=\frac{\sigma}{2\varepsilon}(\sqrt{x^2+R^2}-x) $$
> 求出每一个微小圆环上的电荷量表达式：$dq=\sigma \cdot 2\pi r \cdot dr$

#### 电势差

&emsp;A->B的电势差表示的是由**A点移动到B点**电场力对于单位点电荷**所做的功**

#### 电势梯度求电场强度

$$ \vec{E}=-\nabla U $$

> 一定要注意这里有一个**负号**

### 欧姆定律的微分形式

$$ \vec{J}=\frac{1}{\rho}\vec{E}=\sigma \vec{E} $$

> $\vec{J}$表示电流密度，$\rho$表示电阻率，$\sigma$表示电导率

### 关于电荷密度

&emsp;给定电荷量的面密度$\sigma$，要转换成线密度，只需要确定一个微元$dx$，然后算出另$x$相等时在面上的一条线段长度$l(x)$，然后将三者相乘即可：**$\sigma l(x)dx$**

### 电介质

#### 极化强度矢量

$$ \vec{P}=\frac{\sum_i \vec{p_i}}{\Delta V} $$

#### 束缚电荷密度

$$ \sigma'=P \cos \theta=\vec{e} \cdot \vec{P} $$

> 束缚电荷面密度等于该处极化强度的法向量，$\vec{e}$为平面法向量，由内指向外

#### 电介质中的场强

$$ E=\frac{E_0}{\varepsilon_r} $$

> $E_0$表示原来的场强大小

#### 极化强度矢量关于场强$E$的表达式

$$ \vec{P}=\chi _e \varepsilon_0 \vec{E}=(\varepsilon_r-1)\varepsilon_0 \vec{E}=(\varepsilon_r\varepsilon_0-\varepsilon_0) \vec{E} $$

> $\chi_e$成为电介质的极化率，$\varepsilon=\varepsilon_0\varepsilon_r$称为电介质的介电常数  
> $\vec{E}$全部都应该取某一点实际的电场强度

#### 环路定理

&emsp;环路定理仍然适用，高斯定理要有一定的变化

#### 高斯定理

$$ \oint_S \vec{E} d \vec{S}=\frac{1}{\varepsilon_0}\sum_{S内}{q_o+q'} $$

> $q'$表示极化电荷

同时有：
$$ \sum_{S内}q’=-\oint_S\vec{P}d\vec{S} $$
带入后移项可得：

$$ \oint_S(\varepsilon_o\vec{E}+\vec{P})d\vec{S}=\sum_{S内} {q_0} $$
$$ D=\varepsilon_o\vec{E}+\vec{P} $$
$$ \oint_S \vec{D}\cdot d \vec{S} =\sum_{S内} {q_0} $$

> 注意这边是$\sum_{S内} {q_0}$而不是$q_内/\varepsilon_o$

由$ \vec{P}=(\varepsilon_r\varepsilon_0-\varepsilon_0) \vec{E} $可得：

$$ \vec{D}=\varepsilon_0\varepsilon_r \vec{E}=\varepsilon\vec{E} $$

### 电容器

#### 平行板电容器的电容

$$ C=\frac{\varepsilon_r \varepsilon_0 S}{d} $$

#### 电容计算

&emsp;利用电容的定义公式来计算

$$ C=\frac{Q}{U} $$

&emsp;只要计算出两个电极之间的电势差U关于Q的表达式

### 静电场能量

#### 点电荷系能量

$$ A=\sum_{i=1}^n\frac{1}{2}q_i U_i $$

#### 电容器储能

电容器的能量可以看作把单位电荷从电势低处搬到电势高处外力做的功

$$ A=\int_0^Q \frac{q}{C}dq=\frac{Q^2}{2C} $$

#### 电场的能量

由电容器储能的表达式，以及
$$ C=\frac{\varepsilon S}{d} $$
$$ \Delta U= Ed $$
可得：
$$ A=\frac{1}{2}\varepsilon E^2 \cdot S d $$
其中$Sd$为面积，则单位体积的电场的能量密度就可以表示为
$$ \omega_e=\frac{1}{2}\varepsilon E^2=\frac{1}{2}\vec{E}\cdot\vec{D}  $$

## 静磁学

### 安培定律

$$ \vec{F}=\frac{\mu_0}{4\pi}I_0d\vec{l_0}\times \frac{Id\vec{l}\times \vec{e}}{r^2} $$

### 毕奥-萨伐尔定律

$$ \vec{B}=\frac{\mu_0}{4\pi}\frac{I_0d\vec{l_0}\times\vec{r}}{r^3} $$

#### 载流直导线

$$ B=\frac{\mu_0 I}{4 \pi a}(\cos \theta_1 - \cos \theta_2) $$

#### 圆电流

$$ B=\frac{\mu_0}{2}\frac{I\cdot R^2}{(x^2+R^2)^\frac{3}{2}} $$

> 可以等效于电流全部汇聚在一点产生电场，因为轴线上点到圆环的距离都是相同的，只要之后乘上角度换算成**延轴线方向**的即可
> 在$x=0$时，表示**圆环中心的磁场强度**：
> $$ B_0=\frac{\mu_0 I}{2R} $$
> 对于圆弧圆心的场强，只需要乘上$$ \frac{弧长}{圆周长} $$
> **圆心场强方向都是垂直于电流所在平面的**

### 安倍环路定理

$$ \oint_L B \cdot dl = \mu_0 \sum I_内 $$

&emsp;微分形式：

$$ \nabla \times B =\mu_0 j $$

> 可以用来求电流分布
> 要求表达出平面内各点的磁场强度表达式，然后通过叉乘就能求解各点电流分布表达式

### 运动电荷的磁场

&emsp;对于点电荷运动，有

$$ I=\frac{dq}{dt} $$

&emsp;可以推出

$$ I=nqv\cdot dS $$

&emsp;如果电荷是周期性运动的，那么电荷产生的电流就为：
$$ I=\frac{q}{T} $$

&emsp;之后带入毕奥萨伐公式求解

### 磁矩

> 只要是由两个异性磁极构成的系统都可以是磁偶极子，环形电流属于等效磁偶极子

#### 磁偶极子磁矩

$$ \vec{p}_m = NI S \vec{e}_n =NI\pi R^2 \vec{e}_n $$

#### 磁力矩

$$ \vec{M}=\vec{p}_m \times \vec{B} $$

> 磁力矩是**力矩**，而磁场属于一个定义的量，与力矩无关
> 该公式对于匀强磁场中任意形状的平面均成立，所以都可以通过使用磁矩计算磁力矩

### 安培力

$$ \vec{F}=I\vec{B} \times d\vec{l} $$

### 洛伦兹力

$$ \vec{F} =q\vec{v}\times\vec{B} $$

### 安培力做功

$$ A=I\Delta \Phi_m  $$

## **Python** 历史

`Python` 的创始人为荷兰人吉多·范罗苏姆（`Guido van Rossum`）。1989年的圣诞节期间，吉多·范罗苏姆为了在阿姆斯特丹打发时间，决心开发一个新的脚本解释程序，作为 ABC 语言的一种继承。之所以选中 `Python` 作为程序的名字，是因为他是 BBC 电视剧——蒙提·派森的飞行马戏团（`Monty Python's Flying Circus`）的爱好者。

1991年，第一个 Python 编译器诞生。它是用C语言实现的，并能够调用C语言的库文件。

`Python 2.0` 于 2000 年 10 月 16 日发布，增加了实现完整的垃圾回收，并且支持 `Unicode`。

`Python 3.0` 于 2008 年 12 月 3 日发布，此版不完全兼容之前的 `Python` 源代码。不过，很多新特性后来也被移植到旧的 `Python 2.6/2.7` 版本。

    python

就可以进入 `Python` 解释器的页面。

    python hello_world.py

也会输出 `"hello world!"` 的结果。
